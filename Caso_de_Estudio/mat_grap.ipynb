Caso de estudio 1 - Filtro paso bajos

```python
#with h5py.File(r"datos_filtro\simulados\Filtered_signal.mat", 'r') as archivo:
#    datos_simulados = archivo[list(archivo.keys())[0]][:]
datos_experimentales = scipy.io.loadmat(r"datos_filtro\experimentales\Filtered_signal.mat")
datos_experimentales = datos_experimentales[list(datos_experimentales.keys())[0]]
```

```python
tiempo_datos_simulados=datos_simulados[:,0]
valor_datos_simulados=datos_simulados[:,1]

tiempo_datos_experimentales=datos_experimentales[0]
valor_datos_experimentales=datos_experimentales[1]
```

Señal de salida simulada

```python
plt.figure(figsize=(10, 5))
plt.plot(tiempo_datos_simulados, valor_datos_simulados, label='Señal filtrada')
plt.title('Señal filtrada simulada')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Sesgo [°/ms]')
plt.legend()
plt.grid()
plt.show()
```

Señal de salida experimental

```python
plt.figure(figsize=(10, 5))
plt.plot(tiempo_datos_experimentales, valor_datos_experimentales, label='Señal filtrada')
plt.title('Señal filtrada experimental')
plt.xlabel('Amplitud [V]')
plt.ylabel('Tiempo [ms]')
plt.legend()
plt.grid()
plt.show()
```

Métricas para la comparación de las señales del caso de estudio 1

```python
# Supongamos que tienes dos vectores de datos ya cargados
vector1 = np.array(valor_datos_simulados)  # Datos del primer vector
vector2 = np.array(valor_datos_experimentales)  # Datos del segundo vector

# Calcular el error promedio absoluto (MAE)
mae = np.mean(np.abs(vector1 - vector2))
print(f"Error Promedio Absoluto (MAE): {mae}")

# Calcular la raíz del error cuadrático medio (RMSE)
rmse = np.sqrt(np.mean((vector1 - vector2) ** 2))
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

mse = np.mean((vector1 - vector2) ** 2)
print("Error Cuadrático Medio (MSE):", mse)
```

Caso de estudio 2 - IMU

```python
with h5py.File(r"datos_imu\simulados\Gyroscope_Bias_sim.mat", 'r') as archivo:
    sesgo_simulado_del_giroscopio = archivo[list(archivo.keys())[0]][:]

with h5py.File(r"datos_imu\simulados\Orientation_Error_sim.mat", 'r') as archivo:
    error_de_orientacion_simulado = archivo[list(archivo.keys())[0]][:]

sesgo_del_giroscopio_experimental = scipy.io.loadmat(r'datos_imu\experimentales\Gyroscope_Bias.mat')
sesgo_del_giroscopio_experimental = sesgo_del_giroscopio_experimental[list(sesgo_del_giroscopio_experimental.keys())[0]]
error_de_orientacion_experimental = scipy.io.loadmat(r'datos_imu\experimentales\Orientation_Error.mat')
error_de_orientacion_experimental = error_de_orientacion_experimental[list(error_de_orientacion_experimental.keys())[0]]
```
