sennal de entrada

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los par√°metros del escal√≥n
t_inicio = -0.1  # Tiempo de inicio de la gr√°fica en segundos
t_subida = 0     # Tiempo en el que sube a 1 V
t_fin = 0.5      # Tiempo en el que baja a 0 V
amplitud = 1.0   # Amplitud del escal√≥n
resolucion = 0.001  # Resoluci√≥n temporal en segundos

# Crear el vector de tiempo desde -0.1 hasta 0.5 segundos
t = np.arange(t_inicio, t_fin + resolucion, resolucion)
# Crear el vector de se√±al escal√≥n: comienza en 0, sube a 1 en t=0 y baja a 0 en t=0.5
escalon = np.where(t >= t_subida, amplitud, 0)
#escalon[t >= t_fin] = 0  # Baja a 0 en t=0.5


# Graficar el escal√≥n
plt.figure(figsize=(8, 4))
plt.step(t, escalon, where='post', label='Se√±al de entrada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Voltaje (V)')
plt.title('Se√±al de entrada simulada')
plt.ylim(-0.1, 1.1)  # Ajuste de l√≠mites para visualizar mejor el cambio
plt.xlim(-0.01, 0.51)  # Limitar el eje x para mostrar el rango deseado
plt.grid(True)
plt.legend()
plt.show()
```
