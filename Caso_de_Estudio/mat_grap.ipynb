Plot .MatFiles

```python
import scipy.io
import matplotlib.pyplot as plt

# Load the .mat file
filtered = scipy.io.loadmat('Filtered_signal.mat')
raw = scipy.io.loadmat('Raw_signal.mat')
```

```python
x_filtered = filtered["ans"][0]  # Assuming the first column is x
y_filtered = filtered["ans"][1]  # Assuming the second column is y

x_raw = raw["ans"][0]
y_raw = raw["ans"][1]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_filtered, y_filtered, label='Data Label')  # Add labels and title as necessary
plt.title('Filtered')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.legend()
plt.grid()
plt.savefig('Filtered.pdf', format='pdf', dpi=600)  # Save as PDF with 300 DPI
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_raw, y_raw, label='Data Label')  # Add labels and title as necessary
plt.title('Raw')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.legend()
plt.grid()
plt.savefig('raw.pdf', format='pdf', dpi=600)  # Save as PDF with 300 DPI
plt.show()
```
