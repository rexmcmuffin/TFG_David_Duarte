```python
import h5py
import scipy.io
import matplotlib.pyplot as plt

# Load the .mat file
#Imu_Data = scipy.io.loadmat('IMUFusionSimulinkModel.mat')
with h5py.File("Gyroscope_Bias_sim.mat", 'r') as archivo:
    print("Claves del archivo:", list(archivo.keys()))
    gyroscope_byas_simulated = archivo["gyroscope_bias"][:]

with h5py.File("Orientation_Error_sim.mat", 'r') as archivo:
    print("Claves del archivo:", list(archivo.keys()))
    orientation_error_simulated = archivo["orientation_error"][:]

gyroscope_byas = scipy.io.loadmat('Gyroscope_Bias.mat')
orientation_error = scipy.io.loadmat('Orientation_Error.mat')
```

```python
x_simulated=gyroscope_byas_simulated[:,0]
y1_simulated=gyroscope_byas_simulated[:,1]
y2_simulated=gyroscope_byas_simulated[:,2]
y3_simulated=gyroscope_byas_simulated[:,3]

x_oe_simulated=orientation_error_simulated[:,0]
y1_oe_simulated=orientation_error_simulated[:,1]

```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_simulated, y1_simulated,y2_simulated,y3_simulated, label='Data Label')  # Add labels and title as necessary
plt.title('Gyroscope Bias')
plt.xlabel('Tiempo')
plt.ylabel('Bias')
plt.legend()
plt.grid()
#plt.savefig('Filtered.pdf', format='pdf', dpi=600)  # Save as PDF with 300 DPI
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_oe_simulated, y1_oe_simulated, label='Orientation Error')  # Add labels and title as necessary
plt.title('Orientation Error')
plt.xlabel('Tiempo')
plt.ylabel('Error')
plt.legend()
plt.grid()
#plt.savefig('Filtered.pdf', format='pdf', dpi=600)  # Save as PDF with 300 DPI
plt.show()
```
